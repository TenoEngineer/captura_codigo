# MUDANÇA DE LÓGICA

Foi alterado a lógica de busca da ordem da proposta das obras. Em vez de buscar pelos pdf nas pastas de aprovações que são poucas, encontrar através dos arquivos de relação de material que também possuem esse código e é mais fácil de copiar.

```python
#FUNÇÃO PARA ENCONTRAR A PASTA COM O CAMINHO
def find_past(path_past, name_past):
    if '.' not in name_past: 
        new_path=os.path.join(path_past, name_past)
        new_past=os.listdir(new_path)
    else:
        next
        return path_past, name_past
    return new_path, new_past
```

```python
def pasta_documento(path, *args):
    pastas = args
    pasta=''
    for x in pastas[0]:
        if 'DOCUMENT' in x.upper():
            pasta = x
    if not pasta:
        quebra= True
        return quebra, quebra
    else:
        doc=find_past(path,pasta)
        path=doc[0]
        past=doc[1]
        return path, past
```

```python
def pasta_RMA(path, *args):
    files=args[0]
    arquivo=None
    for file in files:
        if 'MATERIA' in file.upper():
            arquivo=file
    if arquivo == None:
        return True, True
    rma=find_past(path,arquivo)
    path=rma[0]
    past=rma[1]
    return path, past
```

```python
def arquivos_excel(path, file):
    if '.xls' in file:  #Selecionando arquivos derivados do xls
        if 'xls' == file[-3:]: #Não suporta arquivos xls
            xls_to_xlsx(path)
            next
        else:
            cod=None
            if file[0]=='~':
                next
            else:
                excel=load_workbook(os.path.join(path,file)) #Seleciona o arquivo
                excel.active
                aba= excel.sheetnames
                planilha = excel[aba[0]]  #Seleciona a aba
                cod = planilha['I1'].value  #Copia valor

                #Testes para ver se o cod pegou certo
                if len(str(cod)) == 9 or len(str(cod)) == 10:
                    cod = planilha['L2'].value  #Copia valor

                if cod == '(51)-':
                    cod = planilha['L1'].value  #Copia valor

                if not cod:
                    next

                elif len(str(cod)) > 50:  #Caso haja alguma fórmula
                    cod=''

                elif len(str(cod)) > 28:
                    y=(cod.split()[-1]).split(sep='-')
                    if len(y)<2:
                        cod=''
                    else:
                        cod = y[0][:2]+ '_' + y[-2][-3:]   #Novo código

                else:
                    cod=None
            return cod
    else:
        next    
    return
```

```python
def xls_to_xlsx(path):
    o=win32.gencache.EnsureDispatch("Excel.Application")
    #o = win32com.client.Dispatch("Excel.Application")
    o.Visible = False
    input_dir = path
    output_dir = path
    files = glob.glob(input_dir + "/*.xls")
    files_xlsx = glob.glob(input_dir + '/*.xlsx')
    print(files_xlsx)
    for filename in files:
        file = os.path.basename(filename)
        output = (output_dir + '\\' + file).replace('.xls', '.xlsx')
        output=output.replace('[','(')
        output=output.replace(']',')')
        if output in files_xlsx:
            print('Já existe arquivo xlsx, não precisa substituir \n')
            next
        else:
            wb = o.Workbooks.Open(filename)
            wb.SaveAs(output, FileFormat = 51)
            wb.Close(True)
            print('Arquivo salvo como xlsx com sucesso {file} \n'.format(file=output))
            next
```
